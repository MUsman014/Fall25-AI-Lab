```python
import random

def label(total: int) -> str:
    if total % 15 == 0:
        return "Fizz Buzz"
    if total % 3 == 0:
        return "Fizz"
    if total % 5 == 0:
        return "Buzz"
    return str(total)

def fizzbuzz_twist_game(rounds=50, min_val=1, max_val=20):
    print("Welcome! Rules:")
    print(" - Turn 1: check the random number itself.")
    print(" - Turn 2+: check (previous + current).")
    print(" - Answer with 'Fizz', 'Buzz', 'Fizz Buzz', or the number (sum).")
    print(" - One mistake ends the game.\n")

    score = 0
    prev = None

    for r in range(1, rounds + 1):
        curr = random.randint(min_val, max_val)

        total = curr if prev is None else prev + curr
        correct = label(total)

        print(f"Round {r} — Computer -> {curr}")
        user_raw = input("Your answer: ").strip()

        
        low = user_raw.lower().replace(" ", "")
        if low == "fizzbuzz":
            guess = "Fizz Buzz"
        elif low == "fizz":
            guess = "Fizz"
        elif low == "buzz":
            guess = "Buzz"
        else:
            
            try:
                guess_num = int(user_raw)
                guess = str(guess_num)
            except ValueError:
                guess = user_raw  

        if guess.lower() == correct.lower():
            score += 1
            print(f" Correct ({correct}). Score: {score}\n")
            prev = curr
        else:
            print(f" Wrong. Correct was: {correct}")
            print(f"Game Over! Final Score: {score}")
            return

    print(f" You cleared {rounds} rounds! Final Score: {score}")
fizzbuzz_twist_game(30)  

```
